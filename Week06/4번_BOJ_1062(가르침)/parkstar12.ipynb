```python
def solve(words,word,K) :
  max_count=-1

  def back(start,path):
    nonlocal max_count
    # K는 뽑을 수 있는 최대 갯수 word는 단어 리스트 갯수 
    # 고려하는 중인 글자 조합의 갯수가 K이거나 word의 갯수와 동일할 경우
    if len(path) == K or len(path) == len(word) :
      max_count=max(max_count,count(path))
      return
    
    for i in range(start,len(word)):
      path.append(word[i])
      back(i+1,path)
      path.pop()
  
  def count(tmp_word) :
    tmp_word_set = set(tmp_word)
    tmp_word_set.update(['a', 'n', 't', 'i', 'c'])
    return sum(1 for w in words if set(w).issubset(tmp_word_set))
  
  # 뽑을 수 있는 글자 갯수가 0 이거나 word의 갯수가 0 일 경우
  # ex) antaaaaatica 이면 word의 갯수는 0 임 
  if K==0 or len(word)==0:  
    max_count=max(max_count,count([])) # 읽을 수 있는 글자가 있는지 고려
  else :
    back(0,[])
  return max_count

#--------------------------------------#
# 입력
N, K=map(int,input().split()) 

# 가르칠 수 있는 글자 갯수가 5개 미만이면 단어를 하나도 알 수 없음음
if K < 5:
  print(0)
  
else: 
  words=[]
  word=set()
  word.update(['a', 'n', 't', 'i', 'c']) # 기본적으로로 'a', 'n', 't', 'i', 'c' 는 알아야 함함

  for _ in range(N):
      origin_word = input()[4:-4] #anta,tica 제외 글자
      words.append(origin_word) #단어 저장
      word.update(origin_word) # 글자를 set으로 저장해 중복 방지

  for char in ['a', 'n', 't', 'i', 'c']: # anta,tica를 제외한 글자만 고려하기 위해 word에서 제외
      word.remove(char)

  word_list=list(word) # 인덱스 사용하기 위해 list로 변환
  result=solve(words,word_list,K-5) # K-5 ->'a', 'n', 't', 'i', 'c'제외 글자 갯수
  print(result)
```

```python
def solve(words,alphabet,K):
  def back(start,path):
    nonlocal max_count
    if len(path) == K or len(path) == len(alphabet) :
      max_count=max(max_count,count(path))
      return
    for i in range(start,len(alphabet)) :
      path.append(alphabet[i])
      back(i+1,path)
      path.pop()

  def count(path) :
    path_set=set(path)
    path_set.update((['a', 'n', 't', 'i', 'c']))
    return sum(1 for w in words if set(w).issubset(path_set))

  max_count = -1 
  back(0,[])
  return max_count

#--------------------------------
N,K=map(int,input().split())
if K<5 :
  print(0)
  exit()

words=[]
alphabet=set()
alphabet.update((['a', 'n', 't', 'i', 'c']))

for _ in range(N):
  tmp=input()[4:-4]
  words.append(tmp)
  alphabet.update(tmp)

if (K-5)==0 or len(alphabet)==5 :
  print(sum(1 for w in words if set(w).issubset({'a', 'n', 't', 'i', 'c'})))
  exit()

else :
  for i in ['a', 'n', 't', 'i', 'c'] :
    alphabet.remove(i)

  alphabet_list=list(alphabet)
  result=solve(words,alphabet_list,K-5)
  print(result)
```
