```python
from collections import deque
import sys
input = sys.stdin.readline

def bfs(start, direc, queue):
    queue.append(start)  # 시작 지점을 큐에 추가
    visited.add(start)  # 시작 지점을 방문 처리
    # BFS 구현
    while queue:
        cur = queue.popleft()  # 큐에서 현재 위치를 꺼냄
        for dx, dy in direc:  # 상하좌우 방향으로 이동
            # 이동한 위치가 유효한 범위 내에 있고, 물에 잠기지 않았으며, 방문하지 않은 경우
            if 0 <= cur[0] + dx < n and 0 <= cur[1] + dy < n:
                if loc[cur[0] + dx][cur[1] + dy] > rain and (cur[0] + dx, cur[1] + dy) not in visited:
                    queue.append((cur[0] + dx, cur[1] + dy)) 
                    visited.add((cur[0] + dx, cur[1] + dy)) 

n = int(input()) 
loc = []  
maxheight, minheight = 0, 100000000  # 지역의 최대 높이와 최소 높이를 초기화
maxResult = 1  # 안전 영역의 최대 개수를 저장할 변수 (최소 1개는 존재)

# 지역의 높이 정보를 입력받아 loc 리스트에 저장
for i in range(n):
    temp = list(map(int, input().split()))  # 한 줄의 높이 정보를 입력받음
    loc.append(temp)  # loc 리스트에 추가
    for i in temp:  # 현재 줄의 높이 정보를 순회하며 최대/최소 높이 갱신
        if i > maxheight:
            maxheight = i
        if i < minheight:
            minheight = i

# 상하좌우 방향 정의
directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

# 비의 높이를 minheight부터 maxheight-1까지 변화시키며 시뮬레이션
for rain in range(minheight, maxheight):
    notSink = set()  # 물에 잠기지 않은 지역을 저장할 집합
    for i in range(n):
        for j in range(n):
            if loc[i][j] > rain:  # 현재 비의 높이보다 높은 지역만 추가
                notSink.add((i, j))
    
    visited = set()  # 방문한 지역을 저장할 집합
    result = 0  # 현재 비의 높이에서의 안전 영역 개수

    # 물에 잠기지 않은 지역을 순회하며 BFS로 안전 영역 탐색
    for i in notSink:
        if i not in visited:  # 아직 방문하지 않은 지역이라면
            queue = deque()  # BFS를 위한 큐 초기화
            bfs(i, directions, queue)  # BFS 실행
            result += 1  # 안전 영역 개수 증가

    # 현재 비의 높이에서의 안전 영역 개수가 최대값보다 크다면 갱신
    if result > maxResult:
        maxResult = result

print(maxResult)
```
